## Clustering des fleurs d’Iris avec l’algorithme K-Means

```python
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

## Importation du Jeu de données Iris  à l'aide de module sklearn.datasets

```python
iris = load_iris()
list(iris.keys())
```

```python
X = iris.data
y = iris.target

plt.scatter(X[:, 0], X[:, 1], c=y, alpha=0.8)
```

```python
iris.feature_names
```

# Importer kmeans  à l'aide de module sklearn.cluster , puis appliquer l'algorithme sur notre dataset

**K-Means** est un algorithme d'apprentissage non supervisé, l’algorithme n’a pas d’étiquette y. 
Il découvre des patterns en fonction des caractéristiques se trouvant dans features X. 
Dans ce cas **K-Means** ne sait pas que la première fleur Iris de notre jeu de données est de
telle ou telle variante, donc vous n'avez pas besoin de faire le split de dataset en test et en train (il y a pas une phase d'entrainement ) 

```python
from sklearn.cluster import KMeans
```

```python
random_state = 0
```

```python
colors = ['green', 'red','blue','cyan', 'magenta','yellow','black','orange']
```

```python
kmeans = KMeans(random_state=random_state).fit(X)
kmeans = KMeans(n_clusters=2, random_state=random_state).fit(X)
y_pred = kmeans.predict(X)
centroids = kmeans.cluster_centers_
centroids
```

```python
fig, axes = plt.subplots(1,3)
i = 0
for n_clusters in [2,3,8]:

        kmeans = KMeans(n_clusters=n_clusters, random_state=random_state).fit(X)
        y_pred = kmeans.predict(X)
        centroids = kmeans.cluster_centers_
        ax = axes[i]
        
        ax.scatter(X[:, 0], X[:, 1], c=y_pred)
        ax.scatter(centroids[:, 0], centroids[:, 1],marker = "x", s=150, 
                linewidths = 5, zorder = 10, c=colors[:n_clusters])
        ax.set_xlabel(iris.feature_names[0])
        ax.set_ylabel(iris.feature_names[1])
        ax.set_title(f"n_clusters={n_clusters}")
        # ax.legend()
        i+= 1
fig.set_size_inches(18, 5, forward=True)
plt.show()
```

```python
y_pred = KMeans(n_clusters=3, random_state=random_state).fit_predict(X)
centroids = kmeans.cluster_centers_

plt.subplot(221)
plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.scatter(centroids[:, 0], centroids[:, 1], centroids[:, 2], marker = "x", s=150, 
        linewidths = 5, zorder = 10, c=['green', 'red','blue','cyan', 'magenta','yellow','black'])
plt.title("n_clusters=3")
```
