## Créer une image du drapeau  de Russie (le drapeau doit avoir un contour noir).

```python
from PIL import Image, ImageDraw
```

```python
img=Image.new("RGB",(602,452), 'black')

for x in range(1,150):
    for y in range(1,600):
        img.putpixel((y,x),(255,255,255))

for x in range(150,300):
    for y in range(1,600):
        img.putpixel((y,x),(0,0,255))

for x in range(300,450):
    for y in range(1,600):
        img.putpixel((y,x),(255,0,0))

img
```

```python
largeur = 300
hauteur = 160

marge = 10
largeur_drap = largeur-marge
hauteur_drap_line = 50
canvas = Image.new('RGB', (largeur, hauteur), 'black')
img_draw = ImageDraw.Draw(canvas)

shape = [(marge, marge), (largeur_drap, hauteur_drap_line)] 
img_draw.rectangle(shape, outline='black', fill='white')

shape = [(marge, hauteur_drap_line), (largeur_drap, hauteur_drap_line*2)] 
img_draw.rectangle(shape, outline='black', fill='blue')

shape = [(marge, hauteur_drap_line*2), (largeur_drap, hauteur_drap_line*3)]
img_draw.rectangle(shape, outline='black', fill='red')
img_draw.text((100, 10+(hauteur_drap_line/3)), 'Stop the war', fill='green')
# canvas.save('drawn_image.jpg')

canvas
```
