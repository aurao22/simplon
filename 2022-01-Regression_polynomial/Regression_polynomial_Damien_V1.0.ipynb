```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
import numpy as np
import matplotlib.pyplot as plt 
import numpy.random as rnd
import sklearn
```

```python
np.random.seed(42)
m = 100
X = 6 * np.random.rand(m, 1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m, 1)
```

```python
import seaborn as sns
```

```python
plt.scatter(X,y)
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
polyFeat=poly.fit(X)
```

```python
X_poly=polyFeat.transform(X)
```

```python
lr = LinearRegression()
```

```python
lr = lr.fit(X_poly, y)
Y_pred = lr.predict(X_poly)
```

```python
plt.plot(X, y, marker='o', ls='', label='data')
#plt.plot(X, y, ls='--', label='Fonction r√©elle')
plt.plot(X, Y_pred, marker='^',alpha=.5, label='prediction polynomiale')
plt.legend()

ax = plt.gca()
ax.set(xlabel='x data', ylabel='y data')
```

```python
print("coeff:"+str(lr.coef_[0][1::]))
print("intercept :"+str(lr.intercept_[0]))
```

```python
plt.scatter(X, Y_pred, marker='*',linestyle='solid')
plt.scatter(X, y, marker='^')
plt.show()
```

# affichons la pr√©diction et les data

```python

x = np.linspace(-3,3, 40)
z = lr.coef_[0][2]*x**2+lr.coef_[0][1]*x+lr.intercept_[0]
plt.plot(x, z)
plt.scatter(X, y, marker='^')
#plt.scatter(X, Y_pred, marker='*')
```
