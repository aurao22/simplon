```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from os import getcwd
```

```python
# Récupère le répertoire du programme
curent_path = getcwd()
curent_path = curent_path + "\\"
curent_path
```

```python
dataset = pd.read_csv(curent_path+"titanic_dataset_kaggle_train.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.fillna(dataset.mean(), inplace=True)
```

```python
dataset.describe()
```

```python
dataset.columns
```

```python
data = dataset.drop(['PassengerId', 'Name', 'SibSp',
       'Parch', 'Ticket', 'Cabin', 'Embarked'], axis=1)
data.head()
```

```python
data.Sex = data.Sex.map({'male':0, 'female':1})
data.head()
```

```python
X = data.drop(['Survived'], axis=1)
y = data.Survived
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0 )
X_train.head()
```

```python
svclassifier = svm.SVC(kernel='linear', random_state=0)
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
svclassifier.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
```
