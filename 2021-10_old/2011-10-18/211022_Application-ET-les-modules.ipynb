## Boolean Indexing

```python
import numpy as np
A = np.array([[3,2,1], [6,5,4]])
 
# masque booléen les valeurs inférieurs à 5
print("---------")
B = A < 5
print("B",B)

# sous-ensemble filtré par le masque booléen (c-ad récuperez les valeurs qui sont inférieurs à 5)
print("---------")
C = A[A < 5]
print("C",C)

# convertit les valeurs inferieur à sélectionnées en 4
print("---------")
D = A
D[D < 5] = 4
print("D",D)
```
