Une bibliothèque de prêt dispose d'un certain nombre d'<mark>ouvrages</mark>, classés par <mark>rayon</mark> (Littérature, Histoire, Géographie, etc.). Chaque ouvrage est l'oeuvre d'un ou <mark>plusieurs auteurs</mark>, et doit également être référencé selon un certain nombre de <mark>mots-clés</mark>.

Chaque <mark>adhérent</mark> peut <mark>emprunter</mark> jusqu'à <mark>5 livres</mark> en même temps, et dispose d'un certain <mark>délai</mark> passé lequel il doit recevoir des <mark>relances</mark> puis des <mark>pénalités</mark>.

On se place dans quatre cas successifs, de complexité croissante :
- la bibliothèque ne possède qu'un seul exemplaire de chaque ouvrage. Elle enregistre uniquement les emprunts présents (il n'y a pas d'historique des emprunts passés).
- la bibliothèque ne possède qu'un seul exemplaire de chaque ouvrage, mais elle tient un <mark>historique de tous les emprunts</mark> qui ont été effectués.
- la bibliothèque tient un historique... et elle est maintenant susceptible de posséder <mark>plusieurs exemplaires de certains ouvrages</mark>
- en plus de cela, il existe une <mark>bibliothèque centrale</mark> et des <mark>antennes locales</mark>. Chaque antenne possède un <mark>fonds</mark> qui lui est <mark>propre</mark>, et peut de surcroit se procurer <mark>certains ouvrages</mark> auprès de la <mark>bibliothèque centrale</mark> (mais pas d'une autre bibliothèque locale). Tout <mark>adhérent</mark> peut automatiquement emprunter des ouvrages dans <mark>toutes les antennes locales</mark>.

> Etablir le MCD et le MLD adéquats dans les quatre cas.
